```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd

image_folder = r"G:\MyUniversity\lecture\大三下\深度学习\lab1\实验一材料\GTSRB\Training\00000"
```

```python
# 打开 PPM 文件
ppm_file = os.path.join(image_folder, "00000_00000.ppm")
image = Image.open(ppm_file)

# 显示图片
plt.imshow(image)
plt.axis('off')  # 不显示坐标轴
plt.show()
```

```python
image = np.array(image)
# 打印图片的形状和数据类型
print("Image shape:", image.shape)
print("Image dtype:", image.dtype)
```

```python
def cut_image(image, x1, y1, x2, y2):
    image = image[y1:y2+1, x1:x2+1, :]
    return image
    

plt.imshow(cut_image(image, 5, 6, 24, 25))
plt.axis('off')
plt.show()
```

```python
csv_file = os.path.join(image_folder, "GT-00000.csv")
# 读取 CSV 文件
df = pd.read_csv(csv_file, sep=';')
df
```

```python
fig, axes = plt.subplots(4, 4, figsize=(8, 8))
fig.suptitle('Cropped Images: class 0', fontsize=16)
for i, ax in enumerate(axes.flat):
    row = df.iloc[i]
    ppm_file = os.path.join(image_folder, row['Filename'])
    image = Image.open(ppm_file)
    image = np.array(image)
    
    x1, y1, x2, y2 = row['Roi.X1'], row['Roi.Y1'], row['Roi.X2'], row['Roi.Y2']
    cropped_image = cut_image(image, x1, y1, x2, y2)
    
    ax.imshow(cropped_image)
    ax.axis('off')

plt.tight_layout()
plt.show()
```
